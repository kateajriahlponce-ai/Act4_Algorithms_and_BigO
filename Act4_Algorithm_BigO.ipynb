```python
# Step 1: Define an algorithm in pseudocode as comments
# Algorithm to find the largest number in a list
# 1. Assume the first number is the largest
# 2. Compare each number with the current largest
# 3. If a number is larger, update the largest
# 4. Return the largest number
# Step 2: Implement it in Python
numbers = [10, 5, 20, 8, 15]
largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num
print("Largest number:", largest)
```

```python
# Step 1: Implement linear search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
# Step 2: Test the function
data = [3, 5, 2, 4, 9]
index = linear_search(data, 4)
print("Found at index:", index)
```

```python
# Step 1: Implement binary search
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
# Step 2: Test with a sorted list
sorted_data = [1, 3, 5, 7, 9, 11]
result = binary_search(sorted_data, 7)
print("Found at index:", result)
```

```python
# Step 1: Implement bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
# Step 2: Test sorting
to_sort = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(to_sort)
print("Sorted list:", to_sort)
```

```python
# Step 1: Implement selection sort
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
# Step 2: Test and observe performance
sample_data = [29, 10, 14, 37, 13]
selection_sort(sample_data)
print("Sorted list:", sample_data)
```

```python
# Step 1: Implement insertion sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
# Step 2: Test and measure performance
arr = [12, 11, 13, 5, 6]
insertion_sort(arr)
print("Sorted array:", arr)
```

```python
def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
print(fib_iter(10))
```

```python
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n-1) + fib_rec(n-2)
print(fib_rec(10))
```

```python
def bubble_sort(arr):
    n = len(arr)
    comparisons, swaps = 0, 0
    for i in range(n):
        for j in range(0, n-i-1):
            comparisons += 1
            if arr[j] > arr[j+1]:
                swaps += 1
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr, comparisons, swaps
print(bubble_sort([5, 3, 8, 4, 2]))            
```

```python
def selection_sort(arr):
    comparisons, swaps = 0, 0
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        swaps += 1
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr, comparisons, swaps
print(selection_sort([5, 3, 8, 4, 2]))
```

```python
def insertion_sort(arr):
    comparisons, swaps = 0, 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            comparisons += 1
            arr[j + 1] = arr[j]
            swaps += 1
            j -= 1
        arr[j + 1] = key
    return arr, comparisons, swaps
print(insertion_sort([5, 3, 8, 4, 2]))
```

```python
import random

def bubble_sort(arr):
    comps, swaps = 0, 0
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            comps += 1
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swaps += 1
    return comps, swaps

def selection_sort(arr):
    comps, swaps = 0, 0
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            comps += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            swaps += 1
    return comps, swaps

def insertion_sort(arr):
    comps, swaps = 0, 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >=0 and arr[j] > key:
            comps += 1
            arr[j+1] = arr[j]
            swaps += 1
            j -= 1
        arr[j+1] = key
    return comps, swaps

# Example run
arr = [5, 2, 4, 3, 1]  # small array for demo
b = arr.copy()
s = arr.copy()
i = arr.copy()
bubble_result = bubble_sort(b)
selection_result = selection_sort(s)
insertion_result = insertion_sort(i)

# Print results
print("Algorithm   Comparisons   Swaps")
print(f"Bubble      {bubble_result[0]}            {bubble_result[1]}")
print(f"Selection   {selection_result[0]}            {selection_result[1]}")
print(f"Insertion   {insertion_result[0]}            {insertion_result[1]}")
```
